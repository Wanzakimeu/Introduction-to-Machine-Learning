```python
#creating and training a model

#import libraries

import pandas as pd
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_error , r2_score , mean_absolute_error

```

```python
#load data set
client = pd.read_csv('https://bit.ly/2ODZvLCHRDataset')

```

```python
#size of data set
client.shape
```

```python
#first and last records of a dataset
client.head(5)
```

```python
 #data types in the dataset

client.info()
```

```python
#cleaning data set

client.columns = client.columns.str.strip().str.upper().str.replace('?', '').str.replace('_', ' ')
client.head()
```

```python
#checking for missing data
client.isnull().any()
```

```python
client.duplicated().sum()
```

```python
#client.isnull().sum()


client.dropna(how ='any', inplace = True)


client.info()
```

```python
client.shape
```

```python
# dealing with duplicates

#finding no. of duplicates
 
client.duplicated().sum()
```

no duplicates found after dropping all null rows

```python
client.head(5)
```

```python
#train test split
# X is the input ; Y is the output
x = client.iloc[:,6:13]

y = client.iloc[:,13]

```

```python
x.head(2)
```

```python
y.head(2)
```

```python
#spliting data into test & training

# X is the input ; Y is the output

xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size = 0.25, random_state = 12345 )
```

```python
xtrain.head(3)
```

```python
ytrain.head(3)
```

```python
#Train/test Model with decision tree
modelDeTr = DecisionTreeClassifier(random_state=12345)

modelDeTr.fit(xtrain,ytrain)

print('Accuracy',modelDeTr.score(xtest,ytest))
```

```python
from pandas.core.common import random_state
#Train/test Model with randomforest model

modelLR= RandomForestClassifier(random_state=12345, n_estimators=9)

modelLR.fit(xtrain,ytrain)

print('Accuracy',modelLR.score(xtest, ytest))

```

```python
#Summary of Findings and Recommendations

#Random forest gives very good results at 0.9 and decision tree gives results of 0.87, which is equally good.
#using random forest we can predict  with 90% accuracy.
```
